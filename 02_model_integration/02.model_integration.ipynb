```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
print(model)
```

```python
response =  model.invoke("Hello how are you")
response
```

```python
from langchain_groq import ChatGroq

model2 = ChatGroq(model="qwen/qwen3-32b")
response2 = model2.invoke("why do parrots talk?")
response2
```

## streaming and batch


```python
model2.invoke("Write me a 200 words paragraph on artificial intelligence")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on artificial intelligence"):
    print(chunk.text,end="|",flush=True)
```

```python
model.invoke("why do parrots have colorful feathers")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers"):
    print(chunk.text,end="|",flush=True)
```

```python
response3 = model.batch([
    "Why do parrots talk?",
    "How do airplanes fly",
    "What is quantam computing"
])
for response in response3:
    print(response)
```

```python
response4 = model.batch([
    "Why do parrots talk?",
    "How do airplanes fly",
    "What is quantam computing"
], config={
    'max_concurrency': 5
})
```
