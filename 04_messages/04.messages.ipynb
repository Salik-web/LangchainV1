```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("Please tell me what is atificial intelligence")
```

## text prompts

```python
model.invoke("what is langchain")
```

```python
from langchain.messages import SystemMessage,HumanMessage,AIMessage

messages = [
    SystemMessage("You are poetry expert"),
    HumanMessage("Write me a poem on artifical intelligence")
]
response = model.invoke(messages)
response.content
```

```python
sys_msg = SystemMessage("You are helpful coding assistant")

messages = [
    sys_msg,
    HumanMessage("How do i create rest api")
]
response = model.invoke(messages)
print(response.content)
```

```python
system_message = SystemMessage(
    """You are a senior Python developer with expertise in web frameworks.
Always provide code examples and explain your reasoning.
Be concise but thorough in your explanations."""
)

message = [
    system_message,
    HumanMessage("How do i create a rest api")
]
response = model.invoke(message)
print(response.content)
```

```python
## message metadata
human_msg = HumanMessage(
    content="Hello",
    name="Ally",
    id = "msg_!23"
)
response = model.invoke([
    human_msg
])
response
```

```python
from langchain.messages import AIMessage, SystemMessage,HumanMessage

ai_msg = AIMessage("I would be happy to help you with that question!")

message = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Whats 2 + 2?")
]

response = model.invoke(message)
response.content
```

```python
response.usage_metadata
```

```python
from langchain.messages import AIMessage, ToolMessage

ai_message = AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_weather",
        "args":{"location":"San francisco"},
        "id":"call_123"
    }]
)

weather_result = "sunny , 72F"

tool_message = ToolMessage(
    content=weather_result,
    tool_call_id = "call_!23"
) 

message = [
    HumanMessage("Whats the weather in San Francisco?"),
    ai_message,
    tool_message
]

response = model.invoke(message)
response
```
